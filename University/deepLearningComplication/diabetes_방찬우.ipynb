# 문제: 회귀
# 입력: 10
# 출력: 1

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential()
model.add(layers.Dense(6,activation='relu', input_shape=(10,)))
model.add(layers.Dense(1, activation='linear'))

model.summary()
```

```python
model.compile(optimizer='rmsprop',loss='mse',metrics='mae')
```

```python
model.fit(x_input,x_target,epochs=5)
```

```python
print(x_target)
```

```python
# 손실이 너무 커서 전처리가 필요해보인다.
# 타깃 데이터를 전처리하자
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# 타깃 데이터는 1차원이다. fit 하기 위해서는 2차원으로 변환해야 한다.
x_target = x_target.reshape((-1,1))
sc.fit(x_target)
x_target = sc.transform(x_target)
```

```python
print(x_target)
```

```python
model.fit(x_input,x_target,epochs=5)
# 전처리하니 손실이 압도적으로 줄었다.
```
