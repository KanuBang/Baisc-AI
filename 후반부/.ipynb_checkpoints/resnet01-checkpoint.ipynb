```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Subtract, Add, Dense, Conv2D, Input, Concatenate

input1 = Input( shape=(32,32,256) )
conv1 = Conv2D(64, (1,1), padding='same') (input1)
conv2 = Conv2D(64, (3,3), padding='same') (conv1)
conv3 = Conv2D(256, (1,1), padding='same') (conv2)
output = Add() ([input1, conv3])

model = keras.Model(inputs=[input1],
                    outputs=[output])
model.summary()
from tensorflow.keras.utils import plot_model
plot_model(model, to_file='model.jpg', show_shapes=True)
```
