### Input과 output에 s 빼먹음
### Concatenate는 더하기라고 생각
### Add는 input들의 사이즈가 같아야 하는 거 같다. 
####  병합 모델 Add는 사이즈가 같아야 합니다. Add는 입력 데이터의 각 차원의 요소를 더하는 방식으로 작동하기 때문입니다. 따라서 입력 데이터의 각 차원의 크기가 서로 다르면 더하기 연산을 수행할 수 없습니다.
### Substract도 input들의 사이즈가 같아야 한다.

### aux_output = Dense(10) (dense64) -> activation 안 쓸때
### 왠만하면 이미지 처리는 padding='same'
### 입력은 여전히 까먹으면 안 된다.

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Conv2D, Flatten, MaxPooling2D,Concatenate,Add,Subtract
from tensorflow.keras.utils import plot_model
```

```python

input1 = Input(shape=(10,))
dense1 = Dense(10, activation='relu') (input1)
dense2 = Dense(20, activation='relu') (dense1)
dense3 = Dense(10, activation='relu') (dense2)
output = Dense(1, activation='sigmoid') (dense3)
model = Model(inputs=[input1], outputs=[output])

plot_model(model, to_file='adsf.jpg', show_shapes=True)
```

```python
input2 = Input(shape=(64,64,1))
conv2D_1 = Conv2D(32,kernel_size=(3,3),activation='relu') (input2)
conv2D_2 = Conv2D(32,kernel_size=(3,3),activation='relu') (conv2D_1)
max_1 = MaxPooling2D((2,2)) (conv2D_2)
conv2D_3 = Conv2D(16, kernel_size=(3,3),activation='relu') (max_1)
conv2D_4 = Conv2D(16, kernel_size=(3,3),activation='relu') (conv2D_3)
max_2 = MaxPooling2D((2,2)) (conv2D_4)
flatten_1 = Flatten() (max_2)
dense_1 = Dense(10) (flatten_1)
output = Dense(1, activation='softmax') (dense_1)
model = Model(inputs=[input2], outputs=[output])

plot_model(model,to_file='asdf.jpg',show_shapes=True)
```

```python
input1 = Input(shape=(256,))
output1 = Dense(1,activation='sigmoid') (input1)
output2 = Dense(10,activation='softmax') (input1)
model = Model(inputs=[input1], outputs=[output1,output2])

plot_model(model, to_file='jajaja.jpg', show_shapes=True)
```

```python
input1=Input(shape=(128,))
input2=Input(shape=(256,))
output=Concatenate()([input1,input2])
model=Model(inputs=[input1,input2],outputs=[output])
plot_model(model, to_file='bnm.jpg',show_shapes=True)
```

```python
input1=Input(shape=(128,))
input2=Input(shape=(128,))
output=Add()([input1,input2])
model=Model(inputs=[input1,input2],outputs=[output])
plot_model(model, to_file='bnm.jpg',show_shapes=True)
```

```python
input3 = Input(shape=(64,))
input4 = Input(shape=(64,))
output = Subtract()([input3,input4])
model = Model(inputs=[input3,input4], outputs=[output])
plot_model(model,to_file='asdf.jpg',show_shapes=True)
```

```python
main_input = Input(shape=(1024,))
dense128 = Dense(128,) (main_input)

aux_input = Input(shape=(10,))
dense64 = Dense(64,) (dense128)

merged = Concatenate()([aux_input,dense64])
aux_output = Dense(10, ) (dense64)

main_output = Dense(1,activation='sigmoid') (merged)

model = Model(inputs=[main_input,aux_input], outputs=[main_output,aux_output])
plot_model(model,to_file='cvbn.jpg',show_shapes=True)
```
