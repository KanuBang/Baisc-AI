```python
from tensorflow.keras.layers import Input,Conv2D,Dense,MaxPooling2D,Concatenate,Subtract,Add
from tensorflow.keras.models import Model
from tensorflow.keras.utils import plot_model
```

```python
pa = Input(shape=(28,28,192))
conv11_1 = Conv2D(64,kernel_size=(1,1), padding='same') (pa)
conv11_2 = Conv2D(96,kernel_size=(1,1), padding='same') (pa)
conv11_3 = Conv2D(16,kernel_size=(1,1), padding='same') (pa)
maxpool_1 = MaxPooling2D((3,3), strides=1, padding='same') (pa)

conv33 = Conv2D(128,kernel_size=(3,3),padding='same') (conv11_2)
conv55 = Conv2D(32,kernel_size=(5,5),padding='same') (conv11_3)
conv11_1_1 = Conv2D(32, kernel_size=(1,1),padding='same') (maxpool_1)

output = Concatenate()([conv11_1,conv33,conv55,conv11_1_1])

model = Model(inputs=[pa],outputs=[output])
```

```python
plot_model(model,to_file='test1.jpg',show_shapes=True)
```

```python
input1 = Input(shape=(32,32,256))
conv1 = Conv2D(64, kernel_size=(1,1), padding='same') (input1)
conv2 = Conv2D(64, kernel_size=(3,3), padding='same') (conv1)
conv3 = Conv2D(256, kernel_size=(1,1), padding='same') (conv2)
output1 = Add()([input1,conv3])

model1 = Model(inputs=[input1], outputs=[output1])
```

```python
plot_model(model1,to_file='test1.jpg',show_shapes=True)
```

```python
input_a = Input(shape=(224,224,3))
input_b = Input(shape=(224,224,3))
conv = Conv2D(16,kernel_size=(1,1))
conv1 = conv(input_a)
conv2 = conv(input_b)
output = Subtract([conv1,conv2])
model = Model(inputs=[input_a,input_b], output=[output])
plot_model(model,to_file='d')
```
